```python
# @title 2: Install dependencies (uv + CUDA)
# @markdown This takes ~1-2 minutes with uv.
import os

!apt install -y python3-dev unzip 2>/dev/null
!curl -LsSf https://astral.sh/uv/install.sh | sh
os.environ['PATH'] = '/root/.local/bin:' + os.environ['PATH']

os.environ['URVC_CONSOLE_LOG_LEVEL'] = 'WARNING'

# Initialize the core engine
!uv run -q --prerelease if-necessary-or-explicit ./src/ultimate_rvc/core/main.py
!uv add --prerelease if-necessary-or-explicit matplotlib-inline==0.1.7

clear_output()
print('✅ All dependencies installed!')
```

```python
# @title 3: Launch New RVC ❤️
# @markdown  #### Choose a sharing method:

method = 'gradio'  # @param ['gradio', 'ngrok']
ngrok_token = ''  # @param {type:'string'}

prerelease = '--prerelease if-necessary-or-explicit'
entrypoint = './src/ultimate_rvc/web/main.py'

if method == 'gradio':
    !uv run $prerelease $entrypoint --share
elif method == 'ngrok':
    try:
        from pyngrok import ngrok as ng
        ng.set_auth_token(ngrok_token)
        ng.kill()
        tunnel = ng.connect(6969)
        print(f'Ngrok URL: {tunnel.public_url}')
        !uv run $prerelease $entrypoint --listen-port 6969
    except Exception as e:
        print(f'Error starting ngrok: {e}')
```
